```python
# Ejemplo: Calculadora básica con menú y condicionales
# Este ejemplo solicita dos valores al usuario y permite elegir la operación aritmética mediante un menú.

# Captura de valores numéricos
x = int(input('Primer valor: '))
y = int(input('Segundo valor: '))

# Menú de operaciones
print('¿Qué operación desea?:')
print(' 1) Suma')
print(' 2) Resta')
print(' 3) Multiplicación')
print(' 4) División')

# Captura de la opción seleccionada
operacion = int(input())

# Condicionales según la operación elegida
if operacion == 1:
    print(f'La respuesta es: {x + y}')
elif operacion == 2:
    print(f'La respuesta es: {x - y}')
elif operacion == 3:
    print(f'La respuesta es: {x * y}')
elif operacion == 4:
    print(f'La respuesta es: {x / y}')
```

```python
# Ejemplo: Ciclo for con rango personalizado (inicio, fin, incremento)
# Este ejemplo imprime los números del 1 al 10 avanzando de 1 en 1.

for i in range(1, 11, 1):
    print(i)
```

```python
# Ejemplo: Ciclo while con contador incremental
# Este ejemplo muestra cómo usar un while para repetir instrucciones hasta que el contador llegue a 5.

cont = 0  # Inicialización del contador

while cont < 5:
    print(f'El valor es: {cont}')
    # Incremento del contador
    cont = cont + 1
```

Cuarta Parte - **Listas**

```python
# Ejemplo: Cálculo del promedio de una lista de calificaciones
# Este ejemplo muestra dos formas de calcular el promedio:
# 1) Usando un ciclo for con índice
# 2) Usando un ciclo for directo sobre los elementos de la lista

# Lista de calificaciones
calif = [4, 6, 9, 7, 5]

# Forma 1: Usando índice
suma = 0
for i in range(len(calif)):
    # print(f'Posición {i} Calificación: {calif[i]}')  # Línea comentada para mostrar solo el promedio
    suma += calif[i]

prom = suma / len(calif)
print(prom)  # Imprime el promedio

# Forma 2: Iterando directamente sobre los elementos
op = 0
for e in calif:
    op += e

print(f'Promedio: {op / len(calif)}')  # Imprime el promedio
```

Actividad 2. Dada una secuencia de 15 númerops, imprimir la cantidad de números pares, convertir los numeros imparos en pares, e imprimir la lista completa



```python
# Actividad 2: Contar números pares y corregir números impares
# Esta actividad solicita 15 números al usuario, cuenta cuántos son pares,
# convierte los números impares en pares (sumando 1) y muestra la lista original y la lista corregida.

# Inicialización de listas y contador
lista = []          # Lista para almacenar los números ingresados
listaCorregida = [] # Lista donde los impares se convierten en pares
cont = 0            # Contador de números pares

# Captura de 15 números
for i in range(15):
    num = int(input("Value: "))
    lista.append(num)  # Guardar número original en la lista

    # Verificación de paridad y corrección de impares
    if num % 2 != 0:
        listaCorregida.append(num + 1)  # Convertir impar en par
    else:
        listaCorregida.append(num)      # Mantener par
        cont += 1                        # Contar pares

# Impresión de resultados
print("Lista:", lista)
print(f"Cantidad de números pares: {cont}")
print("Lista corregida:", listaCorregida)
```

```python
# Ejemplo: Cálculo de estadísticas específicas de columnas de un DataFrame
# Este ejemplo muestra cómo obtener la media, el valor mínimo y el valor máximo de ciertas columnas.

# Promedio de la columna 'Parcial 1'
print(df['Parcial 1'].mean())

# Valor mínimo de la columna 'Parcial 1'
print(df['Parcial 1'].min())

# Valor máximo de la columna 'Parcial 3'
print(df['Parcial 1'].max())
```

```python
# Ejemplo: Cálculo de promedio por fila y asignación de estatus
# Este ejemplo muestra cómo:
# 1) Calcular el promedio de varias columnas para cada estudiante
# 2) Asignar un estatus según el promedio usando np.where

import numpy as np  # Asegurarse de que numpy esté importado

# Calcular promedio de los tres parciales por fila
df['prom'] = df[['Parcial 1', 'Parcial 2', 'Parcial 3']].mean(axis=1)

# Asignar estatus según el promedio (>=7 Aprobado, <7 Reprobado)
df['estatus'] = np.where(df['prom'] >= 7, 'Aprobado', 'Reprobado')

# Mostrar el DataFrame actualizado
df
```

```python
# Ejemplo: Graficar la evolución de las calificaciones de un alumno específico
# Este ejemplo selecciona las notas de un alumno por su ID y crea un gráfico de línea mostrando su desempeño.

import matplotlib.pyplot as plt  # Asegurarse de que matplotlib esté importado

# Selección del alumno por ID
alumno = 5
notas = df[df['ID'] == alumno][['Parcial 1', 'Parcial 2', 'Parcial 3']].iloc[0]

# Creación del gráfico de línea
plt.plot(['Parcial 1', 'Parcial 2', 'Parcial 3'], notas)  # Eje X: parciales, Eje Y: calificaciones
plt.ylim(0, 10)           # Limitar el eje Y de 0 a 10
plt.ylabel('Calificaciones')
plt.title('Evolución del alumno')
plt.show()  # Mostrar gráfico
```

```python
# Ejemplo: Leer y modificar un archivo Excel con pandas
# Este ejemplo muestra cómo:
# 1) Leer un archivo Excel
# 2) Calcular el promedio de tres parciales por alumno
# 3) Asignar un estatus según el promedio
# 4) Guardar el DataFrame modificado en un nuevo archivo Excel

import pandas as pd
import numpy as np

# Leer archivo Excel
df = pd.read_excel('datos.xlsx')

# Mostrar las primeras filas
df.head()

# Calcular promedio de los tres parciales por fila
df['prom'] = df[['Parcial 1', 'Parcial 2', 'Parcial 3']].mean(axis=1)

# Asignar estatus según el promedio (>=7 Aprobado, <7 Reprobado)
df['estatus'] = np.where(df['prom'] >= 7, 'Aprobado', 'Reprobado')

# Guardar DataFrame modificado en un nuevo archivo Excel
df.to_excel('datos_mod.xlsx', index=False)
```

```python
# Ejemplo: Uso de arrays de NumPy para análisis de datos
# Este ejemplo muestra cómo calcular promedio, valor mínimo y máximo de un array de temperaturas.

import numpy as np  # Asegurarse de importar NumPy

# Array de temperaturas
temps = np.array([20, 23, 27, 25, 23, 19])

# Mostrar el array
print('Temperaturas: ', temps)

# Cálculos estadísticos
print('Promedio: ', temps.mean())  # Promedio de las temperaturas
print('Mínimo: ', temps.min())     # Valor mínimo
print('Máximo: ', temps.max())     # Valor máximo
```

```python
# Ejemplo: Cálculo de potencia eléctrica usando arrays de NumPy
# Este ejemplo muestra cómo multiplicar voltajes y corrientes elemento a elemento para obtener la potencia.

import numpy as np  # Asegurarse de importar NumPy

# Arrays de voltajes y corrientes
voltajes = np.array([5, 5, 5, 5])
corrientes = np.array([1, 2, 3, 4])

# Cálculo de potencia (P = V * I)
potencias = voltajes * corrientes

# Mostrar resultados
print('Potencias: ', potencias)
```
